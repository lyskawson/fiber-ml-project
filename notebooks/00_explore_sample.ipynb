# 00 — Eksploracja danych sample

Wczytuje `data/sample/T35_RH20/Pomiar1.txt` i `Pomiar3.txt`, wyświetla podstawowe statystyki.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

from fiber_ml.ingest.parser import parse_file
from fiber_ml.utils.paths import DATA_SAMPLE_DIR
```

```python
mf1 = parse_file(DATA_SAMPLE_DIR / "T35_RH20" / "Pomiar1.txt")
mf3 = parse_file(DATA_SAMPLE_DIR / "T35_RH20" / "Pomiar3.txt")

print(f"Pomiar1: {mf1.n_points} points, acquired {mf1.acquired_at}")
print(f"Pomiar3: {mf3.n_points} points, acquired {mf3.acquired_at}")
```

```python
mf1.data.describe()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 6), sharex=True)

for ax, mf, label in zip(axes, [mf1, mf3], ["Pomiar1", "Pomiar3"]):
    ax.plot(mf.data["length_1_m"], mf.data["amplitude_db_mm"], lw=0.5)
    ax.set_ylabel("Amplitude (dB/mm)")
    ax.set_title(f"{label} — T=35°C, RH=20%")

axes[-1].set_xlabel("Length (m)")
plt.tight_layout()
plt.show()
```

```python
# Spectral Shift sparsity
ss = mf1.data["spectral_shift_ghz"]
print(f"Non-null Spectral Shift values: {ss.notna().sum()} / {len(ss)}")
print(f"Length range with data: {mf1.data.loc[ss.notna(), 'length_1_m'].min():.4f} — {mf1.data.loc[ss.notna(), 'length_1_m'].max():.4f} m")
```
